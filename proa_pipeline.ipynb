```python
clauses = [
    'Amendments to pleadings must be filed within 15 days of the filing of the initial pleading.',
    'A person who discloses genetic information in violation of this section is liable for a civil penalty of not more than $10,000. The attorney general may bring an action in the name of the state to recover the penalty, plus reasonable attorneys fees and court costs.',
    'A victim of human trafficking, as defined in Section 236.1 of the Penal Code, may bring a civil action for actual damages, compensatory damages, punitive damages, injunctive relief, any combination of those, or any other appropriate relief.',
    'The lessees of any railroad or the company having possession of the railroad shall be liable to an action of any kind in the same court or jurisdiction as were the lessors or owners of the railroad before the lease.',
    'The buyer may recover from the seller as damages the difference between the cost of cover and the contract price together with any incidental or consequential damages.'
]
answers = ['No','No','Yes','Yes','Yes']
```

```python
reasons = []
for i, question in enumerate(clauses):
    completion = palm.generate_text(
        model=model,
        prompt=f"""
        You are a legal expert.
        Clause:{clauses[i]}
        Answer:{answers[i]}
        A private right of action is when a regular person, a private citizen, is legally entitled to enforce their rights under a given statute.
        Give reason for above Clause answer pair.
        Think Step by Step.
        """,
        temperature=0.3,
        max_output_tokens=1024,
    )
    print(completion.result)
    # output = json.loads(completion.result)['reason']
    reasons.append(completion.result)
```

```python
#creating a new prompt template
prompt_template_temp = "A private right of action is when a regular person, a private citizen, is legally entitled to enforce their rights under a given statute. Does the clause specify a private right of action?\n"
for i in range(len(clauses)):
    prompt_template_temp += f'''
Clause: {clauses[i]}
A: {answers[i]}
R: {reasons[i]}
'''
prompt_template_temp+='''
Q: {{text}}
A:
R:
'''
print(prompt_template_temp)
```

```python
test_df = dataset_proa["test"].to_pandas()
prompts = generate_prompts(prompt_template=prompt_template_temp, data_df=test_df)
# print(len(prompts))
print(prompts[0])
```

```python
completion = palm.generate_text(
    model=model,
    prompt=f"""
    You are a legal expert who only gives Yes/No answers with explaination in json format with keys as answer and reason.
    Clause:{prompts[0]}
    Output Format: json
    A private right of action is when a regular person, a private citizen, is legally entitled to enforce their rights under a given statute. Does the clause specify a private right of action? Think Step by Step
    """,
    temperature=0.3,
    # The maximum length of the response
    max_output_tokens=1024,
)
print(completion.result)
output = json.loads(completion.result)['answer']
print(output)
```

```python
generations = []
for i, prompt in enumerate(prompts):
    print(f"Generating for prompt {i}")
    completion = palm.generate_text(
        model=model,
        prompt=f"""
        You are a legal expert who only gives Yes/No answers with explaination in json format with keys as answer and reason.
        Statement:{prompt}
        Output Format: json
        A private right of action is when a regular person, a private citizen, is legally entitled to enforce their rights under a given statute. Does the clause specify a private right of action? Think Step by Step
        """,
        temperature=0.3,
        # The maximum length of the response
        max_output_tokens=1024,
    )
    print(completion.result)
    if(completion.result == None):
        output = "No"
    else:
        output = json.loads(completion.result)['answer']
    generations.append(output)
```
